```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Titanic.csv')
df.head(10)
```

```python
print(df.isnull().sum())
```

```python
df_cleaned_row = df.dropna()
```

```python
df_cleaned_row
```

```python
df_cleaned_col = df.dropna(axis=1)
df_cleaned_col.head(10)
```

```python
df.head(20)
```

```python
print(df['Age'].dtype)
```

```python
df_cleaned_row['Age'] = df_cleaned_row['Age'].astype('int64')
print(df_cleaned_row['Age'].dtype)
```

```python
df_mean = df.fillna(df.mean())
df_mean.head(10)
```

```python
df.describe()
```

```python
df_mode = df.fillna(df.mode().iloc[0])
df_mode.head(10)
```

```python
df_median = df.fillna(df.median())
df_median.head(10)
```

```python
df_cleaned_row['Age'] = pd.to_numeric(df_cleaned_row['Age'])
```

```python
df_cleaned_row['Log_Age'] = np.log(df['Age'])
fig,axes = plt.subplots(1,2,figsize=(10,4))
axes[0].hist(df_cleaned_row['Age'])
axes[0].set_title('Original Age')
axes[1].hist(df_cleaned_row['Log_Age'])
axes[1].set_title('Transformed Age (Logarithmic)')
plt.show()
```

```python
df_cleaned_row['Sqrt_Age'] = np.sqrt(df_cleaned_row['Age'])

plt.figure(figsize=(10,4))
plt.hist(df_cleaned_row['Sqrt_Age'],bins=30)
plt.title('Transformed Square Root of Age Distribution')
plt.xlabel('Square Root of Age')
plt.ylabel('Frequency')
plt.plot()
```

```python
plt.figure(figsize=(10,4))
plt.hist(df_cleaned_row['Age'],bins=30)
plt.title('Original Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.plot()
```
